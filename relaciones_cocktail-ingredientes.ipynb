```python
import pandas as pd
import sqlite3
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Gaspar\Desktop\Coctails_db\all_drinks.csv', delimiter = ",")
```

```python
df_temp = pd.DataFrame()
```

```python
data = {}

for i in np.arange(0,546):
    for j in np.arange(1,16):
        ing = 'strIngredient' + str(j)
        meas = 'strMeasure' + str(j)
        
        if type(df[ing][i]) == str:
            
            data['id_cocktail'] = df['strDrink'][i]
            data['id_ingredient'] = df[ing][i]
            data['measure'] = df[meas][i]
            df_temp = df_temp.append(data, ignore_index=True)
```

```python
def to_lower(text):
    return text.lower()


df_temp['id_ingredient'] = df_temp['id_ingredient'].apply(to_lower)
```

```python
conn = sqlite3.connect('cocktails_database.db')
```

```python
df_cocktail = pd.read_sql_query("SELECT * FROM Cocktail", conn)
```

```python
df_temp = pd.merge(df_temp, df_cocktail, left_on='id_cocktail', right_on='Cocktails')
```

```python
df_temp.head()
```

```python
df_temp = df_temp.drop(columns = ['id_cocktail', 'Instruction', 'id_glass', 'id_category', 'Cocktails'])
```

```python
df_temp.rename(columns={'ID': 'id_cocktail'}, inplace=True)
```

```python
df_temp.head()
```

```python
df_ingredient = pd.read_sql_query("SELECT * FROM Ingredients", conn)
```

```python
df_temp = pd.merge(df_temp, df_ingredient, left_on='id_ingredient', right_on='Ingredients')
```

```python
df_temp.head(100)
```

```python
df_temp = df_temp.drop(columns = ['Ingredients', 'id_ingredient'])
```

```python
df_temp.rename(columns={'ID': 'id_ingredient'}, inplace=True)
```

```python
df_temp.head()
```

```python
orden = ['id_cocktail', 'id_ingredient', 'measure']
df_RelacionesCocktailsIngredients = df_temp[orden]
```

```python
df_RelacionesCocktailsIngredients.to_sql('RelacionCocktail-Ingredients', conn, index=False, if_exists='replace')
```
